### Importamos las librerías necesarias para el análisis

```python
import pandas as pd
import numpy as np
import io
```

```python
from google.colab import files
uploaded = files.upload()
```

```python
df = pd.read_csv(io.BytesIO(uploaded["cancerdata.csv"]))  #creamos df
df
```

```python
# vamos a distinguir variables categóricas de las numéricas...

X = df.iloc[:,2:] #Variables numéricas desde posición 2 en adelante
Y = df.iloc[:,1]  #Variables categóricas ('diagnosis')
```

```python
X.head()  #visualizamos varibales numéricas
```

```python
# Contamos los datos de la variable categórica

Y.value_counts()
```

Si solo tuviera datos B o solo M no serviría este modelo de clasificación porque faltarían observaciones

Creamos división de modelos

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.3, random_state = 4)

modelo = DecisionTreeClassifier()
modelo.fit(X_train, Y_train)
resultado = modelo.score(X_test, Y_test)  # puntuación, precisión entre nuestro modelo de X_test y Y_test
print(resultado)
```

```python
# Validación cruzada k fold

from sklearn.model_selection import KFold

modelo = DecisionTreeClassifier()
kfold_validation=KFold(10)
```

```python
# buscamos que de estas 10folds  muestras sacamos los promedio, para cruzar validaciones (10 validaciones)

from sklearn.model_selection import cross_val_score

resultados = cross_val_score(modelo, X, Y, cv = kfold_validation)
print(resultados)
resultados.mean()
```

Como los valores score son cercanos al 100%, el modelo es muy bueno, incluso el calor medio.
