### Muestreo sistemático:

```python
def systematic_sampling(econdata, step):
  indexes = np.arange(0, len(econdata), step=step) #recorre rango desde 0 hasta len del df econdata, step función de intervalos, lo definiremos
  systematic_sample = econdata.iloc[indexes]  #función se va a definir por lo lcalización donde esté econdata, y se define por los indices
  return systematic_sample

systematic_sample = systematic_sampling(econdata, 3)  #elige cada 3 observaciones (filas) del df
systematic_sample
```

#### Ahora queremos que nuestros datos digan que la proporción es la siguiente:
1) Hoteles en Cuauhtémoc 0.5

2) Museos en Cuauhtémoc 0.2

3) Hoteles en Venustiana Carranza 0.1

4) Mercados en Cuauhtémoc 0.1

4) MErcados en Venustiana Carranza 0.1

```python
# Vamos a crear función para adaptar la selección de la porporción total para que la muestra tenga estas proporciones que deseamos
# para readaptarlos a los datos que tenemos, que sean de 50, 20, 10, 10 , 10 por ciento.
# configuramos todos los datos, creamos una nueva tabla y luego le indicamos como queremos que sea este muestreo estratificado

def data_estratificada(econdata, nombres_columnas_estrat, valores_estrat, prop_estrat, random_state=None):
  df_estrat = pd.DataFrame(columns = econdata.columns) #creamos df solo con nombre de columnas de df econdata

  pos = -1
  for i in range(len(valores_estrat)):
    pos += 1
    if pos == len(valores_estrat) -1:
      ratio_len =len(econdata) - len(df_estrat)
    else:
      ratio_len = int(len(econdata) * prop_estrat[i])

    df_filtrado = econdata[econdata[nombres_columnas_estrat] == valores_estrat[i]]
    df_temp = df_filtrado.sample(replace=True, n=ratio_len, random_state=random_state)

    df_estrat = pd.concat([df_estrat, df_temp])
  return df_estrat
```

```python
valores_estrat = ['Cuautémoc,Hotel', 'Cuautémoc,Museo', 'Venustiano Carranza,Hotel', 'Cuauhtémoc,Mercado','Venustiano Carranza,Mercado']
prop_estrat = [0.5, 0.2, 0.1, 0.1, 0.1]
df_estrat = data_estratificada(econdata, 'estratificado', valores_estrat, prop_estrat, random_state=42)
df_estrat
```
